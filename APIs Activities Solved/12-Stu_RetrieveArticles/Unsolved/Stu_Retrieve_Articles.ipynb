```python
# Dependencies
import requests
from config import api_key
from pprint import pprint

base_url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Store a search term
search_term = "granola"

# Search for articles published between a begin and end date
begin_date = "begin_date=20120101"
end_date = "end_date=20121231"

page = "page=0"

# Build url
url = base_url + "api-key=" + api_key + "&" + begin_date + "&" + end_date + "&q=" + search_term + "&" + page

print(url)
```

```python
# Retrieve articles
data = requests.get(url).json()
```

```python
pprint(data["response"]["docs"])
```

```python
output = []
```

```python
for article in data["response"]["docs"]:
    output.append(f'{article["snippet"]} - ')
```

```python
for article in output:
    print(output)
```
