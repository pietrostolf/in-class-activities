```python
#Dependencies
import requests
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
base_url = "https://api.tvmaze.com/search/shows?q="
```

```python
# data = requests.get(base_url).json()
# data[0]["show"]["rating"]
```

```python
rating = []
```

```python
#list of tv show titles to query
tv_shows = ["Altered Carbon", "Grey's Anatomy", "This is Us", "The Flash", "Vikings", "Shameless", "Arrow", "Peaky Blinders", "Dirk Gently"]

# make iterative requests to TVmaze search endpoint
for show in tv_shows: 
    combined_url = f"{base_url}{show}"
    data = requests.get(combined_url).json()
    print(show)
    print(data[0]["show"]["rating"]["average"])
    rating.append(data[0]["show"]["rating"]["average"])
```

```python
# create dataframe
ratings_dic = {
    "Show" : tv_shows,
    "Rating" : rating, 
}
ratings_dic
```

```python
ratings_df = pd.DataFrame(ratings_dic)
ratings_df.head()
```

```python
# use pandas to create a bar chart from the dataframe
plt.bar(ratings_df['Show'], ratings_df['Rating'], color='green')
plt.title('TV Shows Ratings')
plt.xlabel('Show')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.show()
```
