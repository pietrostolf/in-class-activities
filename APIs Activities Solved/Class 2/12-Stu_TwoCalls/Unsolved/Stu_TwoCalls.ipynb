```python
# Dependencies
import requests

url = "http://api.worldbank.org/v2/"
```

```python
# Get the list of lending types the world bank has
lending_response = requests.get(f"{url}lendingTypes?format=json").json()
lending_response
```

```python
total = int(lending_response[0]["total"])
total
```

```python
lending_response[1][0]["id"]
```

```python
types_list = []
for type in range(0,total):
    types_list.append(lending_response[1][type]["id"])
types_list
```
