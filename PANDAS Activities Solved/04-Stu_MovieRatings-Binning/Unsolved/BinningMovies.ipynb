```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
csv_path = "Resources/movie_scores.csv"
movies_df = pd.read_csv(csv_path)

movies_df.head()
```

```python
# Figure out the minimum and maximum IMDB user vote count
min_rating = movies_df['IMDB_user_vote_count'].min()
max_rating = movies_df['IMDB_user_vote_count'].max()
print(f'{min_rating}, {max_rating}')
```

```python
# Create bins in which to place values based upon IMDB vote count
bins = np.linspace(243, 334164, num=10)


# Create labels for these bins
labels = ["F", "D", "C", "B", "A", "A+", "A++", "A+++", "A++++" ]
```

```python
# Slice the data and place it into bins
movies_df["IMDB User Votes Group"] = pd.cut(movies_df["IMDB_user_vote_count"], bins, labels=labels)
movies_df.head()
```

```python
# Place the data series into a new column inside of the DataFrame
```

```python
# Create a GroupBy object based upon "IMDB User Votes Group"
grouped = movies_df.groupby('IMDB User Votes Group')

# Find how many rows fall into each bin
counts_df = movies_df['IMDB User Votes Group'].value_counts()

# Get the average of each of the first 5 rating columns within the GroupBy object
movies_avg = grouped.mean()

counts_df
movies_avg
```
