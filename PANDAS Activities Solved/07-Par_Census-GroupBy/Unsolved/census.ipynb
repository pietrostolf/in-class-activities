```python
# Dependencies
import pandas as pd
```

```python
# Save file path to variable
file = "Resources/census_data_2016-2019.csv"
```

```python
# Read with Pandas
df = pd.read_csv(file)
df
```

```python
df_1= df[["Year", "State", "Population", "Employed Civilians", "Unemployed Civilians", "People in the Military", "Poverty Count" ]]
df_2 = df[["Year", "State", "Median Age", "Household Income", "Per Capita Income" ]]
df_2
```

```python
# Create a dataframe of the totals for each state by year.
group_by_df1_ave = df_1.groupby(["Year", "State"])
df_tot_final = group_by_df1_ave[["Population", "Employed Civilians", "Unemployed Civilians", "People in the Military", "Poverty Count" ]].sum().head()
df_tot_final
```

```python
group_by_df2_ave = df_2.groupby(["Year", "State"])
df_avg_final = group_by_df2_ave[["Year", "State", "Median Age", "Household Income", "Per Capita Income" ]].mean().head()
df_avg_final
```

```python
df_avg_final.columns
```

```python
# Rename columns to make them more understandable
df_avg_final_ren = df_avg_final.rename(columns={'Household Income': 'Average Household Income', 'Per Capita Income': 'Average Per Capita Income', })
df_avg_final_ren.head()
```

```python
# Export the DataFrames to CSV
df_tot_final.to_csv('Totals.csv', index=True)
df_avg_final_ren.to_csv('Averages.csv', index=True)
```
