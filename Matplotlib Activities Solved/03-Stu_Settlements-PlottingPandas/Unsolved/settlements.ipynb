```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
union_settlements_df = pd.read_csv("Resources/union_settlements_1995.csv")
union_settlements_df.head()
```

```python
# Get total settlements by union
groupby_union = union_settlements_df.groupby(["UNION"])
per_union_df = groupby_union.count()
per_union_df2 = per_union_df[["NEGOTIATION NUMBER"]]
per_union_df2
```

```python
# Configure plot, figsize, title, and axis labels
set_plot = per_union_df2.plot(kind='bar', figsize=(8, 6))
plt.title("Major Collective Barganing Settlements (1995)")
set_plot.set_xticklabels(per_union_df2.index,rotation = 45,ha="right")
set_plot.set_ylabel("number of settlements")
plt.tight_layout()
plt.show()

# Configure x-tick rotation


# Show plot
```
