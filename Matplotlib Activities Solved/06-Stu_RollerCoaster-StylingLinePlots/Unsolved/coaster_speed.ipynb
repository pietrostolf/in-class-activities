```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
danger_drop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
railgun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
# time = [x*10 for x in range(1,14)]
time = np.arange(0, 130, 10)
time
```

```python
plt.plot(time, danger_drop, linewidth=1, marker = "*", color='red', label="Danger Drop")
plt.plot(time, railgun, linewidth=1, marker = "o", color='blue', label="RailGun")
plt.xlabel("Coaster Runtime")
plt.xlim(0, 120)
plt.ylabel("Speed (mph)")
# plt.ylim(0, 50000)
plt.title("Coaster Speed over Time")
plt.legend(loc="lower right")
plt.grid()
plt.savefig("myGraph.png")
plt.show()
```
